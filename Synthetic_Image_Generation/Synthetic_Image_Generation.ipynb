```python
from torch import autocast
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
	"CompVis/stable-diffusion-v1-4",
	use_auth_token=True
).to("cuda")

```

```python
prompt = "A photo of an Indian astronaut riding a futuristic open jeep on Mars"
```

```python
import os

# Generate and save 3 images
for i in range(1, 4):
    with autocast("cuda"):  # Use autocast for better GPU performance
        image = pipe(prompt).images[0]  
        image.show()
        image.save(f"astronaut_rides_car_{i}.png")
        image.show()

```
